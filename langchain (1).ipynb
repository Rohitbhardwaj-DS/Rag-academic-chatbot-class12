```python
from langchain_community.document_loaders import DirectoryLoader, UnstructuredFileLoader
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_text_splitters import CharacterTextSplitter
```

```python
import nltk 
nltk.download("punkt_tab")
```

```python
docs_dir_path="docs_dir"
vector_db_path="vector_db"
collection_name="documnet_collection"
```

```python
loader=DirectoryLoader(path=docs_dir_path,
                       glob="./*.pdf", 
                       loader_cls=UnstructuredFileLoader)
                       
```

```python
documents=loader.load()
```

```python
len(documents)
```

```python
text_splitters=CharacterTextSplitter(chunk_size=2000,chunk_overlap=500)
```

```python
text_chunk=text_splitters.split_documents(documents)
```

```python
embedding = HuggingFaceEmbeddings()
```

```python
vector_store=Chroma.from_documents(documents=text_chunk,
                                   embedding=embedding, 
                                   persist_directory=vector_db_path, 
                                   collection_name=collection_name)
                                   
```

```python
from langchain_classic.chains import RetrievalQA
```

```python
os.environ['GROQ_API_KEY']="GROQ_API_KEY"
```

```python
vector_db_path="vector_db"
collection_name="documnet_collection"
```

```python
llm=ChatGroq(model="llama-3.1-8b-instant",temperature=0.0)
```

```python
vector_store=Chroma(
    collection_name=collection_name,
    embedding_function=embedding,
    persist_directory=vector_db_path)
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k": 3})
```

```python
qa_chain=RetrievalQA.from_chain_type(
    llm=llm,
    chain_type='stuff',
    retriever=retriever,
    return_source_documents=True)
```

```python
query="what does the document say about Adaptive radiation"
response=qa_chain.invoke({"query":query})
print(response["result"])
```
